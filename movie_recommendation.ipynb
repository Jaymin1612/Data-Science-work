```python
import pandas as pd

r_cols = ['userId','movieId','rating']
ratings = pd.read_csv('d:/Study/Data Science course/Data sets/MovieLens/ml-latest-small/ratings.csv', names=r_cols, usecols=range(3))

m_cols = ['movieId','title']
movies = pd.read_csv('d:/Study/Data Science course/Data sets/MovieLens/ml-latest-small/movies.csv', names=m_cols, usecols=range(2))

ratings = pd.merge(movies,ratings)

ratings.head()
```

```python
ratings = ratings.ix[1:]

ratings.head()
```

```python
userRatings = ratings.pivot_table(index='userId',columns='title',values='rating', aggfunc='last')
userRatings.head()
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix = userRatings.corr(method='kendall', min_periods=100)
corrMatrix.head()
```

```python
myRatings = userRatings.iloc[2].dropna()
myRatings
```

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print "Adding sims for " + myRatings.index[i] + "..."
    # Retrieve similar movies to this one that I rated
    sims = corrMatrix[myRatings.index[i]].dropna()
    # Now scale its similarity by how well I rated this movie
    sims = sims.map(lambda x: x * myRatings[i])
    # Add the score to the list of similarity candidates
    simCandidates = simCandidates.append(sims)
    
#Glance at our results so far:
print "sorting..."
simCandidates.sort_values(inplace = True, ascending = False)
print simCandidates.head(10)
```
